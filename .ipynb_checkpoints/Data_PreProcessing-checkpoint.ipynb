```python
df = pd.read_csv("Sale Table.csv")

# Tạo DataFrame từ dữ liệu
df = pd.read_csv("Sale Table.csv")

# Chuyển đổi cột 'Date' thành kiểu dữ liệu ngày
df['DateTime'] = pd.to_datetime(df['DateTime'])

# Tạo biểu đồ đường
plt.figure(figsize=(12,8))
plt.figure(figsize=(12, 5))
plt.plot(df['DateTime'], df['TotalPrice'])

# Đặt tiêu đề và nhãn cho trục x và y
plt.title('DataPreProcessingg')
plt.xlabel('DateTime')
plt.ylabel('TotalPrice')

# Xác định định dạng hiển thị ngày trên trục x
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
# Xác định góc quay cho nhãn trục x
plt.xticks(rotation=45)

plt.show()
```

```python
# Đọc dữ liệu từ file CSV
df = pd.read_csv("Sale Table.csv")

# Chuyển đổi cột 'DateTime' thành kiểu dữ liệu ngày
df['DateTime'] = pd.to_datetime(df['DateTime'])

# Tạo biểu đồ đường và chỉnh kích thước
plt.figure(figsize=(12, 6))
plt.plot(df['DateTime'], df['TotalPrice'])

# Đặt tiêu đề và nhãn cho trục x và y
plt.title('Data Preprocessing')
plt.xlabel('Date Time')
plt.ylabel('Total Price')

# Xác định định dạng hiển thị ngày trên trục x
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

# Xác định góc quay cho nhãn trục x
plt.xticks(rotation=45)

# Điều chỉnh khoảng cách giữa các thành phần trong hình vẽ
plt.subplots_adjust(bottom=0.2)

plt.show()
```
